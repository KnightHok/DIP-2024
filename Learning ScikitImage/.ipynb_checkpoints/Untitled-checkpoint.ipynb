```python
import numpy as np
from matplotlib import pyplot as plt

random_image = np.random.random([500, 500])

plt.imshow(random_image, cmap='gray')
plt.colorbar()
```

```python
from  skimage import data
coins = data.coins()

print('Type:', type(coins))
print('dtype:', coins.dtype)
print('shape:', coins.shape)

plt.imshow(coins, cmap='gray')
```

```python
cat = data.chelsea()
print("Shape:", cat.shape)
print("Values min/max:", cat.min(), cat.max())

plt.imshow(cat)
```

Create a red square by using standard array slicing and manipulation

```python
cat[10:110, 10:110, :] = [255, 0, 0]
plt.imshow(cat)
```

## Displaying images using matplotlib

```python
img0 = data.chelsea()
img1 = data.rocket()
```

```python
f, (ax0, ax1) = plt.subplots(1, 2, figsize=(20, 10))

ax0.imshow(img0)
ax0.set_title("Cat", fontsize=18)
ax0.axis("off")

ax1.imshow(img1)
ax1.set_title('Rocket', fontsize=18)
ax1.set_xlabel(r'Launching position $\alpha=230$')

ax1.vlines([202, 300], 0, img1.shape[0], colors='magenta', linewidth=3, label='Side tower position')
ax1.plot([168, 150, 200], [400, 200, 300], color='white', linestyle='--', label='Side angle')
```
